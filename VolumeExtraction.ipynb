### Code for getting the volume values from the .stats file, from freesurfer

```python
# Population from TOLD project

import os
import csv

#define the output path for the CSV file
output_path = "/home/daniela/Documents/PROJECTS/TOLD/Brain/MRI/freesurfer/"
# Define the output CSV file
OUTPUT_FILE = output_path + "subjects_GM_data.csv"

# Write the CSV header
with open(OUTPUT_FILE, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Subject", "EstimatedTotalIntraCranialVol", "postcentral_lh", "precentral_lh", "postcentral_rh", "precentral_rh"])

    # Loop through subjects sub-01 to sub-05
    for subj in ["01", "02", "03", "04", "05"]:
        # Define the stats file paths for the current subject
        lh_path = f"/home/daniela/Documents/PROJECTS/TOLD/Brain/MRI/freesurfer/sub-{subj}/stats/lh.aparc.DKTatlas.stats"
        rh_path = f"/home/daniela/Documents/PROJECTS/TOLD/Brain/MRI/freesurfer/sub-{subj}/stats/rh.aparc.DKTatlas.stats"

        # Extract EstimatedTotalIntraCranialVol (same for both hemispheres)
        with open(lh_path, 'r') as f:
            for line in f:
                if "EstimatedTotalIntraCranialVol" in line:
                    eTIV = float(line.split()[-2].rstrip(','))
                    break

        # Extract postcentral GrayVol for left and right hemispheres
        with open(lh_path, 'r') as f:
            for line in f:
                if "postcentral" in line:
                    postcentral_lh = line.split()[3]
                    break

        with open(rh_path, 'r') as f:
            for line in f:
                if "postcentral" in line:
                    postcentral_rh = line.split()[3]
                    break

        # Extract precentral GrayVol for left and right hemispheres
        with open(lh_path, 'r') as f:
            for line in f:
                if "precentral" in line:
                    precentral_lh = line.split()[3]
                    break

        with open(rh_path, 'r') as f:
            for line in f:
                if "precentral" in line:
                    precentral_rh = line.split()[3]
                    break

        # Write the data to the CSV file
        writer.writerow([f"sub-{subj}", eTIV, postcentral_lh, precentral_lh, postcentral_rh, precentral_rh])

print(f"Data extraction complete. Results saved to {OUTPUT_FILE}")
```
